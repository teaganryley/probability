## Coin Toss Experiment

This notebook uses a simple Python script to demonstrate the coin toss experiment. It is a classic random experiment in probability, relayed by Mandelbrot in <em>The Misbehavior of Markets (2006)</em>, as seen in <em>An Introduction to Probability Theory and its Applications (1950)</em> by Feller. 

```python
import matplotlib.pyplot as plt
import numpy as np

n = 10000
winnings_h = []
total = 0

ct_series = np.random.choice(2, n)

for i in ct_series:
    if i > 0:
        total += 1
        winnings_h.append(total)
    else:
        total -= 1
        winnings_h.append(total)

plt.figure(num=1,figsize=(12.8,9.6))           
plt.plot(winnings_h, 'ro')
plt.autoscale(enable=True)
plt.show()
```
